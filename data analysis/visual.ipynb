```python
plt.figure(figsize=(15,6), dpi= 100) # dpi dots per inch
x = np.linspace(0,10,100)
y = np.random.randn(100)
plt.plot(x,y, linewidth= 1, color='blue', linestyle=':', marker='s', markersize=5) #linewidth for the boldness of plot line
#marker ='o'....cicle bn jayega
#'s' = squares
```
